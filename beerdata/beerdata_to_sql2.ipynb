```python
jsonfiles = ['smoke.json', 'fringe.json', 'eastforty.json']

Breweries = ["Smoke", "Fringe", "East Forty"]
```

```python
def JSONtoDataFrame(JSONFile, beername):
    # Open JSON Files
    with open(JSONFile) as f:
        data = json.load(f)

    # Converts data to DataFrame
    beerDF = pd.DataFrame(data[beername])
    # beerDF.head(10)

    return beerDF
```

```python
# Goes Through DataFrame and Removes Beers not Brewed by Company
def RemoveWrongBeers(beerDF, beername):
    for index, row in beerDF.iterrows():
        if not row['name'].startswith(beername):
            beerDF = beerDF.drop([index])
        else:
            #print('next')
            continue
    return beerDF
```

```python
beer_df = JSONtoDataFrame(jsonfiles[0], Breweries[0])
beer_df = RemoveWrongBeers(beer_df, Breweries[0])
beer_df.insert(0, "brewery", Breweries[0])

for x in range(len(Breweries) - 1):    
    b2 = JSONtoDataFrame(jsonfiles[x+1], Breweries[x+1])
    b2 = RemoveWrongBeers(b2, Breweries[x+1])
    b2.insert(0, "brewery", Breweries[x+1])
    beer_df = beer_df.append(b2)

beer_df.head()
```
