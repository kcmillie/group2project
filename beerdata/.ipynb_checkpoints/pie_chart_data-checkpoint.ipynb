```python
link = "year2.js"
with open(link) as f:
    data = json.load(f)

# Converts data to DataFrame
yearDF = pd.DataFrame(data)

yearDF.head()
```

```python
lookup = []
for index,row in yearDF.iterrows():
    brewery = {}
    brewery['name'] = row['Name']
    brewery['rating'] = row['Google Rating']
    brewery['year'] = row['Year Opened']
    lookup.append(brewery)
print(lookup)

a = {'name': 'Fringe', 'rating': 4.8, 'year': 2016}
b = {'name': 'Smoke', 'rating': 4.3, 'year': 2016}
```

```python
for index,row in beerDF.iterrows():
    name = row['brewery']
    for x in range(len(lookup)):
        if (lookup[x]['name'] == name):
            beerDF.loc[index,'google rating'] = lookup[x]['rating']
            beerDF.loc[index,'year founded'] = lookup[x]['year']
beerDF.head(500)
```

```python
BeerCount = {
    'no info': 0,
    'Water Beer': 0,
    'Normal Beer': 0,
    'Boozy Beer': 0,
    'Very Boozy': 0
}

BeerRating = {
    'no rating': 0,
    'Bad Beer': 0,
    'Good Beer': 0,
    'Great Beer': 0
}

def p2f(x):
    return float(x.strip('%'))

def stripScore(x):
    return float(x.strip('/5'))

for index,row in beerDF.iterrows():
    if(row['abv'] == '-'):
        BeerCount['no info'] += 1
    elif (p2f(row['abv']) < 5.0):
        BeerCount['Water Beer'] += 1
    elif (p2f(row['abv']) < 7.0):
        BeerCount['Normal Beer'] += 1
    elif (p2f(row['abv']) < 9.0):
        BeerCount['Boozy Beer'] += 1
    else:
        BeerCount['Very Boozy'] += 1
        
    if(row['score'] == '-/5'):
        BeerRating['no rating'] += 1
    elif(stripScore(row['score']) < 3.0):
        BeerRating['Bad Beer'] += 1
    elif(stripScore(row['score']) < 4.0):
        BeerRating['Good Beer'] += 1
    else:
        BeerRating['Great Beer'] += 1

print(BeerCount)
print(BeerRating)
```
