```python
import pandas as pd
import numpy as np

import json
from pprint import pprint
import matplotlib.pyplot as plt

import seaborn as sb
```

```python
jsonfiles = ['smoke.json', 'fringe.json', 'eastforty.json']

Breweries = ["Smoke", "Fringe", "East Forty"]
```

```python
def JSONtoDataFrame(JSONFile, beername):
    # Open JSON Files
    with open(JSONFile) as f:
        data = json.load(f)

    # Converts data to DataFrame
    beerDF = pd.DataFrame(data[beername])
    # beerDF.head(10)

    return beerDF
```

```python
# Goes Through DataFrame and Removes Beers not Brewed by Company
def RemoveWrongBeers(beerDF, beername):
    for index, row in beerDF.iterrows():
        if not row['name'].startswith(beername):
            beerDF = beerDF.drop([index])
        else:
            #print('next')
            continue
    return beerDF
```

```python
def FindMeanColumn(columnName):
    a = pd.to_numeric(columnName, errors = 'coerce')
    return(a.mean())
```

```python
BeerData = []

#beer_df = JSONtoDataFrame(jsonfiles[3], Breweries[3])
#beerDF = RemoveWrongBeers(beer_df, Breweries[3])
#beerDF.head(25)

for x in range(len(Breweries)):
    brewery = {}
    beer_df = JSONtoDataFrame(jsonfiles[x], Breweries[x])
    beer_df = RemoveWrongBeers(beer_df, Breweries[x])
    beer_df.head()

    # remove "Ratings" from column 1 & find average score
    beer_df['# of ratings'] = beer_df['# of ratings'].str.split(' ').str[0]
    total_ratings = pd.to_numeric(beer_df['# of ratings'], errors = 'coerce').sum()

    #getting score average
    beer_df['score'] = beer_df['score'].str.split('/').str[0]
    averageScore = FindMeanColumn(beer_df['score'])

    # remove % from column & find average abv
    beer_df['abv'] = beer_df['abv'].str.strip('%')
    averageABV = FindMeanColumn(beer_df['abv'])
    
    # average calories
    averageCalories = FindMeanColumn(beer_df['calories'])
    
    #average IBU
    averageIBU = FindMeanColumn(beer_df['ibu'])


    brewery['Name'] = Breweries[x]
    brewery['Average Score'] = averageScore
    brewery['Total Ratings'] = total_ratings
    brewery['Average ABV'] = averageABV
    brewery['Average Calories'] = averageCalories
    brewery['Average IBU'] = averageIBU
    BeerData.append(brewery)
    
```
